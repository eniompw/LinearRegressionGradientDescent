```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data: 30 noisy points around y = 2x + 50
np.random.seed(42)
n, noise_std = 30, 8
X = np.arange(n)
y = 2 * X + 50 + np.random.normal(0, noise_std, n)

# Initialise parameters
W, b, lr = 0.0, 0.0, 0.001
step, epochs = 100, 10_000
hist_W, hist_b = [], []

# Gradient descent
for epoch in range(epochs):
    if epoch % step == 0:
        hist_W.append(W); hist_b.append(b)
    y_pred = W * X + b
    W -= lr * -(2/n) * np.sum(X * (y - y_pred))
    b -= lr * -(2/n) * np.sum(y - y_pred)

hist_W.append(W); hist_b.append(b)
print(f"W: {W:.3f}, b: {b:.3f}")

# Plot setup
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(X, y, color='steelblue', label="Noisy data")
line, = ax.plot([], [], 'r', lw=2, label="Prediction")
ax.set(xlim=(X.min()-1, X.max()+1), ylim=(y.min()-10, y.max()+10))
ax.legend()

# Animation
def animate(i):
    line.set_data(X, hist_W[i] * X + hist_b[i])
    ax.set_title(f"Epoch: {i * step}")
    return line,

ani = FuncAnimation(fig, animate, frames=len(hist_W), interval=50, blit=True)
ani.save('linreg_gd_anim.gif', writer='pillow', fps=20)
plt.close()
```
