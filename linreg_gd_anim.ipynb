```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data: 30 noisy points around y = 2x + 50
n_samples = 30
X = np.arange(n_samples)

np.random.seed(42)                 # optional: reproducible noise
noise_std = 8                      # try 3, 5, 8, 12...
noise = np.random.normal(0, noise_std, size=n_samples)

y = 2 * X + 50 + noise             # noisy targets

# Parameters and learning rate
W = 0.0
b = 0.0
learning_rate = 0.001
epochs = 10_000

history_W = []
history_b = []
step_size = 100

for epoch in range(epochs):
    if epoch % step_size == 0:
        history_W.append(W)
        history_b.append(b)

    y_pred = W * X + b
    dW = -(2 / n_samples) * np.sum(X * (y - y_pred))
    db = -(2 / n_samples) * np.sum(y - y_pred)
    W -= learning_rate * dW
    b -= learning_rate * db

history_W.append(W)
history_b.append(b)

print(f"[Linear] W: {W:.3f}, b: {b:.3f}")

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(X, y, color='blue', label="Noisy data")
line, = ax.plot([], [], 'r', linewidth=2, label="Prediction")

ax.set_xlim(X.min() - 1, X.max() + 1)
ax.set_ylim(y.min() - 10, y.max() + 10)
ax.legend()

def animate(i):
    w_i = history_W[i]
    b_i = history_b[i]
    y_pred_i = w_i * X + b_i
    line.set_data(X, y_pred_i)
    ax.set_title(f"Epoch: {i * step_size}")
    return line,

ani = FuncAnimation(fig, animate, frames=len(history_W), interval=50, blit=True)

# Save the animation as a GIF (requires the 'pillow' library: pip install pillow)
ani.save('gradient_descent.gif', writer='pillow', fps=20)

plt.close() # Prevents displaying a static duplicate plot
```
