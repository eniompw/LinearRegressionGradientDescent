```python
# Importing Libraries
import numpy as np

n = 30              # number of data points
# Data y = 2x + 50
x = np.arange(n)		# 200 causes an overflow in training
y = 2 * x + 50
# Add Gaussian noise to data
noise = np.random.normal(0, 5, n)
y = y + np.round(noise)

# Print first 10 values
print("x:", x)
print("y:", y)
```

```python
# Draw all Data points
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
theta = np.polyfit(x, y, 1)		# poly degree i.e. linear
plt.axline((0, theta[1]), slope=theta[0])
```

```python
# Draw all Data points
import matplotlib.pyplot as plt
plt.scatter(x,y)

w = 0.1
b = 0.01
pred = (w * x) + b
plt.plot(x,pred,'r')
```

```python
# Draw all Data points
plt.scatter(x,y)

w = 0.1
b = 0.01
pred = (w * x) + b
plt.plot(x,pred,'r')

dx = [x[15], x[15]]
dy = [y[15], pred[15]]
plt.plot(dx, dy, 'g')
```

```python
lr = 0.0001
costs = []

for i in range(50000):
  # Calculating the current cost
  costs.append(np.mean((y - pred) ** 2))

  # Calculating the gradients
  dw = -(2/n) * sum(x * (y - pred))
  db = -(2/n) * sum(y - pred)

  # Updating weights and bias
  w = w - lr * dw
  b = b - lr * db
  pred = (w * x) + b

print("weight:" , w)
print("bias:" , b)
```

```python
# Visualising the learning
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.plot(costs[50:])  # first 50 has steap drop
```

```python
# Final Model vs Data
plt.plot(x,pred,'r')
plt.scatter(x,y)
```
